Використовується набір penguins з бібліотеки seaborn.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

1. Попередній аналіз даних:

Завантажити датасет.

```python
df = sns.load_dataset('penguins')
```

Вивести перші 10 рядків таблиці.

```python
df.head(10)
```

Перевірити пропущені значення.

```python
df.isnull().sum()
```

2. Статична візуалізація з Seaborn

Boxplot: маса тіла пінгвінів (body_mass_g) у розрізі виду (species).


```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='species', y='body_mass_g', data=df, color='royalblue')
plt.title('Маса тіла пінгвінів e розрізі виду')
plt.xlabel('Вид')
plt.ylabel('Маса тіла')
plt.show()
```

Pairplot: побудувати pairplot для числових змінних, кольором виділити sex.

```python
sns.pairplot(df, hue='sex', diag_kind='hist', palette='viridis')
plt.suptitle('Pairplot числових змінних за статтю', y=1.02)
plt.show()
```

Countplot: кількість пінгвінів по островах (island).

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='island', data=df, palette='viridis')
plt.title('Кількість пінгвінів по островах')
plt.xlabel('Острів')
plt.ylabel('Кількість')
plt.show()
```

3. Інтерактивна візуалізація з Plotly

Scatter plot: зв’язок між довжиною та шириною дзьоба (bill_length_mm vs bill_depth_mm) з кольором по species і підписами осей.


```python
fig = px.scatter(df, x='bill_length_mm', y='bill_depth_mm', color='species',
                 title='Зв’язок між довжиною та шириною дзьоба з розподілом на види',
                 labels={'bill_length_mm': 'Довжина дзьоба', 'bill_depth_mm': 'Ширина дзьоба'},
                 color_discrete_sequence=px.colors.qualitative.Prism)
fig.show()
```

Box plot (інтерактивний): маса тіла (body_mass_g) по статі (sex), кольором по species.

```python
fig = px.box(df, x='sex', y='body_mass_g', color='species', title='Маса тіла по статі',
             labels={'sex': 'Стать', "body_mass_g": 'Маса тіла'},
             color_discrete_sequence=px.colors.qualitative.Prism)
fig.show()
```
